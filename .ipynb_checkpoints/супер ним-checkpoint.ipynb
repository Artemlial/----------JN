```python
import numpy as np
from numpy.random import randint as rr

letters = 'abcdefgh'

pole = np.array( [['_' for i in range(8)] for j in range(8)])

'''заполняем поле'''

for k in range(10):
    i = rr(0,8,1)
    j = rr(0,8,1)
    pole[i,j] = '@'

'''функция для вывода поля в удобном для чтения виде'''

def fldprnt(fld):
    k = 1
    print('\n')
    for i in fld:
        print(k,'| ',end='')
        k +=1
        print(*i,sep=' | ',end=' |')
        print('\n')

    print('    ',end='')
    print(*letters,sep='   ')

fldprnt(pole)

'''check ingame status'''

def konets(arr):
    if '@' not in arr:
        print(f'\nИгра окончена, победил игрок номер {2 - acts%2}')
        return True
    return False

'''check if dvizh is valid'''

def norm(stroka):
#     if isinstance(stroka,str):
        if len(stroka) > 1:
            print('please enter 1 symbol: a-h for columns and 1-8 for lines!')
            return False
        elif str.isalpha(stroka):
            if stroka not in letters:
                print('letter must be from "a" to "h"')
                return False
            return True
        elif str.isdecimal(stroka):
            if int(stroka) < 1:
                print('number must be in range [1,8]')
                return False
            return True
        else:
            print('Please enter valid command')
            return False

'''игровой цикл'''

acts = 0    

while not konets(pole):
    dvizh = input('please enter command\n').strip()
    while not norm(dvizh):
        dvizh = input('\n').strip()
    if str.isalpha(dvizh):
        pole[:,letters.index(dvizh)] = '_'
    else:
        pole[int(dvizh)-1,:] = '_'
    acts += 1
    fldprnt(pole)
```
