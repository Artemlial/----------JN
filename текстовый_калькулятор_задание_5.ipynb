```python
def calc(strk: str) -> str:

    '''создаем словарь dec_t для перевода входной строки в числа'''

    decimals = 'ноль один два три четыре пять шесть семь восемь девять десять одиннадцать двенадцать тринадцать четырнадцать пятнадцать шестнадцать семнадцать восемнадцать девятнадцать'.split()
    n10_to_90 = 'двадцать тридцать сорок пятьдесят шестьдесят семьдесят восемьдесят девяносто сто'.split()
    operations = 'плюс минус умножить'.split()
    dec_t = dict(zip(decimals,[i for i in range(20)]))     
    dec_t.update(zip(n10_to_90,[i for i in range(20,101,10)]))

    '''создаем словарь rev_dec_t для обратного перевода результата операции в слова'''

    n200_to_1000 = 'двести триста четыреста пятьсот шестьсот семьсот восемьсот девятьсот тысяча'.split()
    rev_dec_t = {v:k for k,v in dec_t.items()}
    rev_dec_t.update(zip([i for i in range(200,1001,100)],n200_to_1000))
    
    '''дополняем словарь dec_t для возможности переводить операции'''
    
    dec_t.update(zip(operations,['+','-','*']))

    '''сплитим строку по пробелам, чтобы можно было перевести по словам'''
    strk = strk.split()
    
    '''удаляем "жировое" слово "на", для упрощённого перевода операции умножения'''
    
    for i in strk:
        if i == 'на':
            strk.remove(i)
            
    '''создаем две переменные, для сбора строки так, чтобы ее могла прочитать функция eval(),
    в b будут собираться числа, в s2 операции и строчные представления чисел'''
    
    b = 0
    s2 = ''

    for i in strk:
        if i not in operations:
            b += dec_t[i] #в этой строке я собираю слова вида "двадцать два" в одно число
        elif b == 0: 
            s2 += dec_t[i] # если в выражении есть два арифметических оператора подряд, благодаря данному условию между ними не будут влезать лишние нули
        else:
            s2 = s2+str(b) + dec_t[i] #если в выражении мы наткнулись на оператор, тогда запихиваем строчное представление b и оператор в будущий аргумент eval() и обнуляем b для будущих сборок
            b = 0 
    if b != 0:
        s2 += str(b) #вводимая строка всегда заканчивается числом, соответственно и то, 
                     #что мы передадим в eval(), должно заканчиваться числом       
    a = eval(s2) 

    sign = a//abs(a) #сохраняем отдельно знак результата вычислений, это понадобится, 
    a = abs(a) #чтобы потом лаконично описать вывод с минусом
    
    try:                                             # если нам свезло, и результат совпадает с одним из ключей 
        return 'минус '*((1-sign)//2) + rev_dec_t[a] # перевода, вернем его сразу
    except KeyError:                                 # остальные случаи разберем в блоке обработки исключения KeyError
        p = 1
        intermed = []
        res=''
        if a > 1000:
            ths,a = a//1000,a%1000

            if ths%100 in [i for i in range(5,20)]:# оцениваем количество тысяч в результате выполнения всех арифметических 
                res +=' тысяч '                    # операций, чтобы собрать выходную строку согласно правилам русского языка
            elif ths%10 == 1:
                res += ' тысяча '
            elif ths%10 > 1 and ths%10<5:
                res +=' тысячи '
            else:
                res +=' тысяч '
            if ths == 2:
                res = 'две' + res
            elif ths == 1:
                res = res[1:]                      # пусть выводит не "одна тысяча", а просто "тысяча" и без лишних пробелов
            else:
                res = rev_dec_t[ths] + res
        if (k := a%100) in rev_dec_t.keys() and k>0: # доп проверка, чтобы числа типа 117 выходили как "сто семнадцать",  
            a -= k                                   # а не, например, "сто один семь"
            intermed.append(rev_dec_t[k])
            p = 3
        while a > 0:
            if (k := a%10**p)>0:
                intermed.append(rev_dec_t[a%10**p]) # переводим результат операции поразрядно
            a -= k
            p += 1
         
        return ('минус '*((1-sign)//2) + res + ' '.join(intermed[::-1])).strip()
```

```python
calc(input('Введите математическое выражение:\n'))
```
